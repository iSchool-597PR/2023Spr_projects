```python
url = "https://www.monster.com/jobs/search?"
params={'q': 'data scientist', 'where': 'united states', 'page': '1'}
en_url = url + urllib.parse.urlencode(params)
en_url
```

```python
response = requests.get(en_url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
soup.find_all('a')
```

```python
url_list = []
```

```python
base_url = "https://www.monster.com/jobs"

while len(url_list)<10:
#     print(len(soup.find_all('a', class_='job-cardstyle__JobCardTitle-sc-1mbmxes-2')))
    for job_url in soup.find_all('a', class_='job-cardstyle__JobCardTitle-sc-1mbmxes-2'):
        print(job_url)
        print(str(job_url.get('href')))
        url_list.append(str(job_url.get('href')))
    load_more = soup.find('a', id="job-search-load-more")
    url_new = base_url + str(load_more.get('href'))[1:]
#     print(url_new)
    resp = requests.get(url_new)
    soup = BeautifulSoup(resp.text, 'html.parser')
```
