```python
import numpy as np
import pandas as pd
import os
```

```python
def read_csv(folder_path):
    files = os.listdir(folder_path)
    df_list = []
    for file in files:
        file_path = os.path.join(folder_path, file)
        df = pd.read_csv(file_path)
        country = file.split('_')[0]
        df['Country'] = country
        df_list.append(df)

    combined_df = pd.concat(df_list, ignore_index = True)
    return combined_df
```
